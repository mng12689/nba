```python
def process_team_data(df):
    # TODO: some seasons are less than 82 games. exclude those for now until we have time to normalize data from those seasons
    df = df[df["GP"] == 82]
    X = df.drop(["TEAM_ID", "TEAM_NAME", "W","L","W_PCT",'CFPARAMS','CFID',"Season","NET_RATING", "GP"],axis=1)
    # TODO: drop rank cols in favor of normalized cols to account for trends across seasons(rank assumes rank-1 and rank+1 are equally spaced)
    X = X[X.columns.drop(list(X.filter(regex='_RANK')))]
    return (X, df["W_PCT"])
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# most seasons are 82 games
# for now throw out the 2 seasons that are less
# eventually we might want to normalize and use the data
#norm_stat = 82/agg["GP"]["mean"]  _stat["min"]/agg["GP"]["mean"]
def graph_time_trends(df):
    # TODO: some seasons are less than 82 games. exclude those for now until we have time to normalize data from those seasons
    df = df[df["GP"] == 82]
    df = df.drop(["TEAM_ID", "TEAM_NAME", "W","L",'CFPARAMS','CFID', "GP"],axis=1)
    # TODO: drop rank cols in favor of normalized cols to account for trends across seasons(rank assumes rank-1 and rank+1 are equally spaced)
    df = df[df.columns.drop(list(df.filter(regex='_RANK')))]
    agg = df.groupby("Season").agg(["min", "mean", "max"])

    n=1
    cols = 5
    rows = len(df.columns)/5

    plt.figure(figsize=(30,100))
    plt.subplots_adjust(hspace=.5)

    for stat_key in agg.columns.get_level_values(0).unique():
        stat = agg[stat_key]
        
        plt.subplot(rows,cols,n)
        plt.title(stat_key)
        plt.plot(agg.index,stat["min"])
        plt.plot(agg.index,stat["mean"])
        plt.plot(agg.index,stat["max"])
        plt.xticks(rotation='vertical')
        
        n=n+1

def graph_win_corr(df):
    df = df[df["GP"] == 82]
    df = df.drop(["TEAM_ID", "TEAM_NAME", "W","L",'CFPARAMS','CFID', "GP"],axis=1)
    # TODO: drop rank cols in favor of normalized cols to account for trends across seasons(rank assumes rank-1 and rank+1 are equally spaced)
    df = df[df.columns.drop(list(df.filter(regex='_RANK')))]
    
    n=1
    cols = 5
    rows = len(df.columns)/5
    
    plt.figure(figsize=(30,100))
    plt.subplots_adjust(hspace=.5)

    for stat_key in df.columns:
        stat = df[stat_key]
        
        plt.subplot(rows,cols,n)
        plt.title(stat_key)
        plt.plot(df["W_PCT"],stat, 'o')
        plt.xticks(rotation='vertical')
        
        n=n+1
    
def graph_all(df):
    graph_time_trends(df)
    #graph_win_corr(df)
    
graph_all(df)
```

```python
def train_ridge_regression(df):
    from sklearn.linear_model import Ridge
    from sklearn.model_selection import GridSearchCV
    from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures
    
    # tried PolynomialFeatures after scaling, but performs worse
    X, y = process_team_data(df)
    X_scaled = MinMaxScaler().fit_transform(X)
    s = GridSearchCV(Ridge(), param_grid={"alpha": [.001, .01,.1,1,10,100]})
    s.fit(X_scaled,y)
    print("Score: ", s.best_score_)
    print("Params: ", s.best_params_)

print("Ridge Regression:")
train_ridge_regression(df)
print("Ridge Regression (absent PLUS_MINUS):")
train_ridge_regression(df.drop(["PLUS_MINUS"],axis=1))
```

```python
def train_lasso_regression(df):
    from sklearn.linear_model import Lasso
    from sklearn.model_selection import GridSearchCV
    from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures
    
    # tried PolynomialFeatures after scaling, but performs worse
    X, y = process_team_data(df)
    X_scaled = MinMaxScaler().fit_transform(X)
    s = GridSearchCV(Lasso(), param_grid={"alpha": [.00001,.0001, .001, .01]})
    s.fit(X_scaled,y)
    print("Score: ", s.best_score_)
    print("Params: ", s.best_params_)
    print("Coefficients: ", dict(zip(X.columns, s.best_estimator_.coef_)))

print("Lasso Regression:")
train_lasso_regression(df)
print("Lasso Regression (absent PLUS_MINUS):")
train_lasso_regression(df.drop(["PLUS_MINUS"],axis=1))
```

```python
def train_decision_tree(df):
    from sklearn.tree import DecisionTreeRegressor
    from sklearn.model_selection import GridSearchCV
    
    X, y = process_team_data(df)
    s = GridSearchCV(DecisionTreeRegressor(), param_grid={"min_samples_leaf": [1,2, 4, 9,25,36]})
    s.fit(X,y)
    print("Score: ", s.best_score_)
    print("Params: ", s.best_params_)
    print("Coefficients: ", dict(zip(X.columns, s.best_estimator_.feature_importances_)))

print("Decision Tree Regression:")
train_decision_tree(df)
print("Decision Tree Regression (absent PLUS_MINUS):")
train_decision_tree(df.drop(["PLUS_MINUS"],axis=1))
```

```python
def train_random_forest(df):
    from sklearn.ensemble import RandomForestRegressor
    from sklearn.model_selection import GridSearchCV
    
    X, y = process_team_data(df)
    s = GridSearchCV(RandomForestRegressor(), param_grid={"min_samples_leaf": [1,2, 4, 9,25,36], "max_features": [.2,4,6,8]})
    s.fit(X,y)
    print("Score: ", s.best_score_)
    print("Params: ", s.best_params_)
    print("Coefficients: ", dict(zip(X.columns, s.best_estimator_.feature_importances_)))

print("Random Forest Regression:")
train_random_forest(df)
print("Random Forest Regression (absent PLUS_MINUS):")
train_random_forest(df.drop(["PLUS_MINUS"],axis=1))
```
